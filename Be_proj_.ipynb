```python
import tensorflow as tf

# The path we just verified
data_dir = '/content/final_dataset/PlantVillage'

# 1. Load Training Data (80% of the photos)
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(224, 224), # Resize everything to be uniform
    batch_size=32
)

# 2. Load Validation Data (20% to "test" the AI)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

class_names = train_ds.class_names
print(f"✅ Success! Found {len(class_names)} diseases/crops.")
```

```python
# 1. Load the pre-trained base model
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False, # We'll add our own classification layer
    weights='imagenet'
)
base_model.trainable = False # Freeze the "old knowledge"

# 2. Add our "New Layers" for disease detection
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(len(class_names), activation='softmax') # The "Final Vote" layer
])

# 3. Compile (The Training Rules)
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
# This will take about 2-5 minutes depending on the GPU
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

```python
model.save('crop_disease_model.h5')
print("✅ Model saved! You can see 'crop_disease_model.h5' in the left sidebar. Download it now!")
```

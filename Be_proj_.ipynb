```python
import tensorflow as tf
import os

# 1. Define the path to your extracted data
data_dir = '/content/crop_data'

# 2. Load the data and split it: 80% for training, 20% for testing (validation)
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(224, 224), # Standard size for MobileNetV2
    batch_size=32
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

# 3. Print the class names (the diseases the AI found)
class_names = train_ds.class_names
print(f"✅ Found {len(class_names)} classes: {class_names}")
```
